# Feature Engineering - Telecom Churn Prediction

On average it costs a company more than 05 times more to aquire a new customer than to retain an existing customer. 

Telcom companies spend huge amonuts and energy to recruit new subscibers,and for some reasons, some customers will leave THE network. This becomes a problem, as Telecom companies ussually incur huge costs to attract subscribers. 

We going to performed __Feature Engineering__ as part of the steps towards building and deploying a churn prediction model, based on the dataset from the BigML website : https://bigml.com/user/francisco/gallery/dataset/5163ad540c0b5e5b22000383

### 2.) Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.drop('Churn',axis = 1)
```

```python
y = dataset['Churn']
```

```python
#from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    stratify=y, 
                                                    test_size=0.2,random_state=42)
```

```python
trainset = pd.concat([X_train,y_train],axis = 1)
```

### Feature Scaling

```python
feature_scale = [feature for feature in X_train.columns if feature not in ['State','Area code','Account length','International plan','Voice mail plan','Customer service calls']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train[feature_scale])
```

```python
data = pd.concat([trainset[['State','Area code','Account length','International plan','Voice mail plan','Customer service calls']].reset_index(drop=True),
                        pd.DataFrame(scaler.transform(trainset[feature_scale]), columns = feature_scale)],
                                axis=1)
```
